```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans

# Step 2: Load built-in (x, y) dataset
X, y = make_moons(n_samples=300, noise=0.1, random_state=0)

# Step 3: Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("make_moons Dataset (x, y)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Step 4: Apply Elbow Method to find optimal clusters
inertia = []
K_values = range(1, 11)

for k in K_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Step 5: Plot Elbow Curve
plt.plot(K_values, inertia, 'bo-')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia / SSE")
plt.show()

# Step 6: Choose optimal k (around 2 for make_moons)
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Step 7: Visualize the final clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids')
plt.title("K-Means Clustering Result on make_moons Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
