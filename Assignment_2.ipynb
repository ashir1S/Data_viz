```python
# Assignment-2: Pandas
import pandas as pd

# -----------------------------
# Q1. Create a dataset
# -----------------------------
data = {
    'ID': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Name': ['Naveen', 'Ritika', 'Abhay', 'Tara', 'Yug', 'Sana', 'Harsh', 'Aarohi', 'Ira'],
    'Age': [22, 24, 23, 25, 21, 26, 27, 23, 22],
    'Department': ['Data', 'HR', 'Finance', 'Data', 'Sales', 'Finance', 'Marketing', 'Data', 'HR'],
    'Salary': [40000, 42000, 39000, 41000, 37000, 43000, 44000, 45000, 46000]
}
df = pd.DataFrame(data)
print("Q1. Created DataFrame:")
print(df)

# -----------------------------
# Q2. Locate rows 0, 4, 7, 8
# -----------------------------
print("\nQ2. Rows 0, 4, 7, 8:")
print(df.loc[[0, 4, 7, 8]])

# -----------------------------
# Q3. Read a CSV file and display it
# -----------------------------
# Example (replace path with your own CSV)
# csv_df = pd.read_csv("sample.csv")
# print(csv_df.head())

print("\nQ3. (Please read your CSV file using pd.read_csv('filename.csv'))")

# -----------------------------
# Q4. Understanding Attribute Types
# -----------------------------
nominal = ['Name', 'Department']
binary = ['Gender (Male/Female)', 'Pass/Fail']
ordinal = ['Performance (Low < Medium < High)']
numeric = ['Age', 'Salary']

print("\nQ4. Attribute Types:")
print("Nominal:", nominal)
print("Binary:", binary)
print("Ordinal:", ordinal)
print("Numeric:", numeric)

# -----------------------------
# Q5. DataFrame Navigation
# -----------------------------
print("\nQ5. 1) Rows 3 to 7:")
print(df.loc[3:7])

print("\nQ5. 2) Rows 4 to 8, Columns 2 to 4:")
print(df.iloc[4:9, 2:5])

print("\nQ5. 3) All rows, Columns 1 to 3:")
print(df.iloc[:, 1:4])

# -----------------------------
# Q6. Read the Iris dataset (from Kaggle)
# -----------------------------
# iris = pd.read_csv("Iris.csv")  # ensure the file path is correct
# print("\nQ6. First 5 rows of Iris dataset:")
# print(iris.head())

print("\nQ6. (Download and read the Iris dataset using pd.read_csv('Iris.csv'))")

# -----------------------------
# Q7. Delete row 4 and column 3 from Iris
# -----------------------------
# iris_modified = iris.drop(index=4).drop(iris.columns[3], axis=1)
# print("\nQ7. After deleting row 4 and column 3:")
# print(iris_modified.head())

print("\nQ7. (Perform deletion using drop: df.drop(index=4).drop(df.columns[3], axis=1))")

# -----------------------------
# Q8. Employee dataset
# -----------------------------
employees = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

emp_df = pd.DataFrame(employees)
print("\nQ8. Employee DataFrame:")
print(emp_df)

# a) Shape
print("\n(a) Shape:", emp_df.shape)

# b) Summary info
print("\n(b) Info:")
print(emp_df.info())

# c) Descriptive statistics
print("\n(c) Describe:")
print(emp_df.describe())

# d) First 5 and last 3 rows
print("\n(d) First 5 rows:")
print(emp_df.head())
print("\nLast 3 rows:")
print(emp_df.tail(3))

# e) Calculations
print("\n(e) Calculations:")
print("Average Salary:", emp_df['Salary'].mean())
print("Total Bonus:", emp_df['Bonus'].sum())
print("Youngest Age:", emp_df['Age'].min())
print("Highest Rating:", emp_df['Rating'].max())

# f) Sort by Salary descending
print("\n(f) Sorted by Salary (descending):")
print(emp_df.sort_values(by='Salary', ascending=False))

# g) Add Performance Category
def category(r):
    if r >= 4.5:
        return 'Excellent'
    elif r >= 4.0:
        return 'Good'
    else:
        return 'Average'

emp_df['Performance'] = emp_df['Rating'].apply(category)
print("\n(g) With Performance Category:")
print(emp_df[['Name', 'Rating', 'Performance']])

# h) Missing values
print("\n(h) Missing Values:")
print(emp_df.isnull().sum())




























# i) Rename column
emp_df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print("\n(i) Renamed Column:")
print(emp_df.head())

# j) Filters
print("\n(j) More than 5 years experience:")
print(emp_df[emp_df['Years_of_Experience'] > 5])

print("\nBelong to IT Department:")
print(emp_df[emp_df['Department'] == 'IT'])

# k) Add Tax column (10% of Salary)
emp_df['Tax'] = emp_df['Salary'] * 0.10
print("\n(k) Added Tax Column:")
print(emp_df[['Name', 'Salary', 'Tax']])

# l) Save to CSV
emp_df.to_csv('employees_modified.csv', index=False)
print("\n(l) Saved modified DataFrame to employees_modified.csv")
```
